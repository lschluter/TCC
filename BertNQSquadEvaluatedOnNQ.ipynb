```python
from datasets import load_dataset

nq_dataset = load_dataset('lucadiliello/naturalquestionsshortqa', cache_dir="/content/cache")
```

```python
from transformers import pipeline
import numpy as np

# Load the question answering pipeline with the "bert-base-uncased" model

qa_pipeline = pipeline("question-answering", model="leonardoschluter/bert-base-uncased-finetuned-nq-finetuned-squad", tokenizer="leonardoschluter/bert-base-uncased-finetuned-nq-finetuned-squad")

# Initialize variables to store evaluation results
total_exact_match = 0
total_f1_score = 0
total_examples = len(nq_dataset["validation"])
```

```python
import datetime
# Loop through the validation dataset
iterator = 0
e = datetime.datetime.now()
print("%s" % e)
for example in nq_dataset["validation"]:
    question = example["question"]
    context = example["context"]
    answers = example["answers"]

    # Use the QA pipeline to get an answer
    predicted_answer = qa_pipeline(question=question, context=context)
    # Calculate exact match (EM) and F1-score for this example
    em = int(predicted_answer["answer"].lower() == answers[0].lower())
    f1 = 2 * (len(set(predicted_answer["answer"].lower().split()) & set(answers[0].lower().split())) / (len(predicted_answer["answer"].lower().split()) + len(answers[0].lower().split())))
    # Update total scores
    total_exact_match += em
    total_f1_score += f1
    iterator = iterator + 1
    if(iterator % 100 == 0):
      e = datetime.datetime.now()
      print("%s" % e)
      print(iterator)
      # Calculate average EM and F1 scores
      average_exact_match = total_exact_match / total_examples
      average_f1_score = total_f1_score / total_examples
      print(f"Average Exact Match (EM): {average_exact_match:.2%}")
      print(f"Average F1 Score: {average_f1_score:.2%}")

# Calculate average EM and F1 scores
average_exact_match = total_exact_match / total_examples
average_f1_score = total_f1_score / total_examples

print(f"Average Exact Match (EM): {average_exact_match:.2%}")
print(f"Average F1 Score: {average_f1_score:.2%}")
```
